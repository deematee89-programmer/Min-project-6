```python
"""
Day 30 Activity: EDA Exercise
Tasks:
1) Load dataset
2) Run univariate, bivariate, and correlation analysis
3) Write 3 insights
"""

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# TODO: Load data from data/day30_eda.csv
# df = pd.read_csv(...)

# TODO: Perform basic EDA and write insights

path="/kaggle/input/datasets/deemass/ds-min-project/day30_eda.csv"
df=pd.read_csv(path)

#=========== Univariate Analysis =============
print("**"*80)
print("Univariate Analysis")
print("**"*80)
print(df.head())
print("**"*80)
print("\nStatistical Summary:\n")
print("**"*80)
print(df.describe())
print("**"*80)
df.hist(figsize=(10,8))
plt.show()

#=========== Correlation Analysis ============
print("**"*80)
print("Correlation Analysis")
print("**"*80)
numeric_cols = df.select_dtypes(include="number").columns
corr_ex = df[numeric_cols].corr()
sns.heatmap(corr_ex, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("EDA Exercise â€” Correlation Heatmap")
plt.show()

#=========== Bivariate Analysis ============
print("**"*80)
print("Bivariate Analysis")
print("**"*80)
sns.scatterplot(data=df, x="age", y="income")
plt.show()
```
